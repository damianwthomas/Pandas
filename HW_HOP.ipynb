```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "Resources/purchase_data.csv"
```

```python
purchasedata_df = pd.read_csv(file_to_load)
summaryPurchase_df=purchasedata_df.groupby (['SN','Gender','Age'], as_index=False)['Price'].sum()
```

```python
totalPlayers=summaryPurchase_df['SN'].count()
totalPlayers_df = pd.DataFrame({'Total Players': totalPlayers},index=[0])
totalPlayers_df
```

```python
numberOfUniqueItems=purchasedata_df['Item ID'].nunique()
averagePrice=purchasedata_df['Price'].mean()
numberOfPurchases=purchasedata_df['Purchase ID'].count()
totalRevenue=purchasedata_df['Price'].sum()

purchaseAnalysis_df = pd.DataFrame({
    'Number of Unique Items': numberOfUniqueItems,
    'Average Price': "${:.2f}".format(averagePrice),
    'Number of Purchases': numberOfPurchases,
    'Total Revenue': "${:.2f}".format(totalRevenue)
}, index=[0], columns=[
    'Number of Unique Items',
    'Average Price',
    'Number of Purchases',
    'Total Revenue'
])
purchaseAnalysis_df
```

```python
males = len(summaryPurchase_df.loc[summaryPurchase_df['Gender'] == 'Male'])
females = len(summaryPurchase_df.loc[summaryPurchase_df['Gender'] == 'Female'])
other = len(summaryPurchase_df.loc[summaryPurchase_df['Gender'] == 'Other / Non-Disclosed'])

genderDemo_df = pd.DataFrame({
    'Total Count' : [ males, females, other],
    'Percentage of Players' : ["{:.2f}".format((males/totalPlayers)*100), "{:.2f}".format((females/totalPlayers)*100),"{:.2f}".format((other/totalPlayers)*100)]  
}, index=['Male','Female','Other / Non-Disclosed'])

genderDemo_df
```

```python
#Get Average
genderMean_df = purchasedata_df.groupby('Gender').mean()
malePPMean = genderMean_df['Price']['Male']
femalePPMean = genderMean_df['Price']['Female']
OtherPPMean = genderMean_df['Price']['Other / Non-Disclosed']

#Get Total Purchase
genderTotal_df = purchasedata_df.groupby('Gender').sum()
malePPTotal = genderTotal_df['Price']['Male']
femalePPTotal = genderTotal_df['Price']['Female']
OtherPPTotal = genderTotal_df['Price']['Other / Non-Disclosed']

#Average Total Per Person
avgTotalMale = malePPTotal/males
avgTotalFemale= femalePPTotal/females
avgTotalOther = OtherPPTotal/other

paGender_DF =  pd.DataFrame({
    'Purchase Count': [females, males, other],
    'Average Purchase Price': ["{:.2f}".format(femalePPMean), "{:.2f}".format(malePPMean), "{:.2f}".format(OtherPPMean)],
    'Total Purchase Value': [femalePPTotal, malePPTotal, OtherPPTotal],
    'Avg Total Purchase per Person': ["{:.2f}".format(avgTotalMale), "{:.2f}".format(avgTotalFemale), "{:.2f}".format(avgTotalOther)]
}, index=['Male','Female', 'Other / Non-Disclosed'], columns=[
    'Purchase Count',
    'Average Purchase Price',
    'Total Purchase Value',
    'Avg Total Purchase per Person'
])

paGender_DF
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]
bin_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'] 
summaryPurchase_df['binAge'] = pd.cut(summaryPurchase_df['Age'], bins, labels=bin_names)
summaryPurchase_df_bin=summaryPurchase_df .groupby (['binAge'], as_index=False)['SN'].count()
summaryPurchase_df_bin['Percent Of Players']=(summaryPurchase_df_bin['SN']/totalPlayers*100).round(2)
summaryPurchase_df_bin.rename(columns = {'SN':'Total Count'}, inplace = True)
summaryPurchase_df_bin
```

```python
purchasedata_df['bin_age'] = pd.cut(purchasedata_df['Age'], bins, labels=bin_names)

purchase_count = purchasedata_df.groupby(['bin_age']).count()['Price']
avg_purchase_price = purchasedata_df.groupby(['bin_age']).mean()['Price'] 
total_purchase_value = purchasedata_df.groupby(['bin_age']).sum()['Price']
person_count=summaryPurchase_df["binAge"].value_counts()
avg_total_per_person = total_purchase_value/person_count

age_demo = pd.DataFrame({'Purchase Count' : purchase_count, 'Average Purchase Price' : (avg_purchase_price).round(2), 'Total Purchase Value': total_purchase_value, 'Average Purchase Total per Person by Age Group' : (avg_total_per_person).round(2)})

age_demo
```

```python
purchase_count = purchasedata_df.groupby(['SN']).count()['Price']
avg_purchase_price = purchasedata_df.groupby(['SN']).mean()['Price'] 
total_purchase_value = purchasedata_df.groupby(['SN']).sum()['Price']

top_spenders_df = pd.DataFrame({'Purchase Count' : purchase_count, 'Average Purchase Price' : (avg_purchase_price).round(2), 'Total Purchase Value': total_purchase_value})
sort_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)
sort_df.head()
```

```python
purchase_count = purchasedata_df.groupby(['Item ID', 'Item Name']).count()['Price']
avg_purchase_price = purchasedata_df.groupby(['Item ID', 'Item Name']).mean()['Price'] 
total_purchase_value = purchasedata_df.groupby(['Item ID', 'Item Name']).sum()['Price']

item_df = pd.DataFrame({'Purchase Count' : purchase_count, 'Average Purchase Price' : (avg_purchase_price).round(2), 'Total Purchase Value': total_purchase_value})
item_sort_df = item_df.sort_values("Purchase Count", ascending=False)
item_sort_df.head()
```
